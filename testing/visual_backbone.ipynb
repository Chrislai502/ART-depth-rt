```python
import torch
from torchvision import models, transforms
from PIL import Image
import requests

# Load the pre-trained ShuffleNet_V2_X0_5 model with specific weights
weights = models.ShuffleNet_V2_X0_5_Weights.IMAGENET1K_V1
model = models.shufflenet_v2_x0_5(weights=weights)
model.eval()  # Set the model to evaluation mode

# Print the model architecture
print(model)

# Use the inference transforms provided by the weights
preprocess = weights.transforms()

# Load a sample image from the internet or local file
url = "https://example.com/sample-image.jpg"  # Replace with an actual URL or path to an image
response = requests.get(url)
img = Image.open(response.raw).convert("RGB")

# Apply the preprocessing transform
input_tensor = preprocess(img)
input_batch = input_tensor.unsqueeze(0)  # Create a mini-batch as expected by the model

# Check the input shape
print(f"Input shape: {input_batch.shape}")

# Forward pass: Compute the model output
with torch.no_grad():
    output = model(input_batch)

# Check the output shape
print(f"Output shape: {output.shape}")

# If you want to get the class probabilities
probabilities = torch.nn.functional.softmax(output[0], dim=0)

# Print the top 5 probabilities along with their corresponding class indices
top5_prob, top5_catid = torch.topk(probabilities, 5)
for i in range(top5_prob.size(0)):
    print(f"Class ID: {top5_catid[i]}, Probability: {top5_prob[i].item()}")
```
